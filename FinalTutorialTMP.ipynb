<h3 style="text-align: center;">Step 2: Data Processing</h3>

The normalization preserves rough ratios of values <i>within</i> an attribute (e.g. double the value for double the population within the available precision - except for extreme values (all values more than 3 SD above the mean are normalized to 1.00; all values more than 3 SD below the mean are normalized to 0.00)). &nbsp;</p>
<h4 style="text-align: center;">Part 1: Tidying data</h4>

<p style="text-align: center;">When data is sparse, e.g. there are a plethora of missing values, we want to organize the data is such a way as to make it more dense. This is called melting the data. For this project, we will melt the data later in order to make a prediction about this data set using a machine learning algorithm. For now, we will sort the values of Violent Crimes Per Community, population of the comunity, and household size. This will sort the data  by communities by the most violent crime, population, and household size at the top of the data frame. We can sort the  data in ascending or descending order. </p>
<br>
<h4 style="text-align: center;">Part 2. Missing Data</h4>
<p style="text-align: center;">This dataframe initially contains 1994 rows. If we drop every row with a "?" to represent missing data from the dataframe, we only end up with 79 rows. We could have tried to impute the data, which means we would add a represented/typical value where these missing points occur. However, many rows contain the same amount of missing values; therefore, imputing the data is too inaccurate of a estimate.</p>

```python
data = data.sort_values(ascending=False, by=["ViolentCrimesPerPop","householdsize","population"])


for index,row in data.iterrows():
     if("?" in str(row)):
            data.drop(index, inplace=True)

data.head(15)
```
