```python
#Finding the mean and median for certain features.
popDensMean = data['PopDens'].mean()
print("The mean population density for the communities in this dataset is: " + str(popDensMean))

popDensMed = data['PopDens'].median()
print("The median population density for the communities in this dataset is: " + str(popDensMed))

print("\n")

popDensMean = data['PctRecImmig10'].mean()
print("The mean of the percentage of immigigrants within the last 10 yers in this dataset is: " + str(popDensMean))

popDensMed = data['PctRecImmig10'].median()
print("The median of the percentage of immigigrants within the last 10 years in this dataset is: " + str(popDensMed))


print("Here we have a histogram showing the variance of the percent of unemployed individuals")
pd.DataFrame.hist(data, column=['PctUnemployed'])
plt.xlabel("Percent Unemployed")
plt.ylabel("# of Communities")
plt.title("Unemployment Histogram")
plt.show()


```
